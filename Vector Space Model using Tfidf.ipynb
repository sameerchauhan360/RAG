```python
# loading the libraries
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
# download the punkt package from nltk
nltk.download('punkt')
nltk.download('stopwords')
```

## Tokenization

```python
# Sample text
text = "Elephants are the largest land animals on Earth. The Pacific Ring of Fire is an area known for frequent earthquakes and volcanic activity."
```

```python
# Tokenize into sentence
sentences = nltk.sent_tokenize(text)
print(f'The Tokenized Sentence are: {sentences}')
```

```python
# Tokenize into words
words = nltk.word_tokenize(text)
print(f'The Tokenized Word are: {words}')
```

## Data Preprocessing

```python
# Process the text: tokenization, only alphanumeric, lower case, no stopwords
def process_text(text):
    processed_text = text.lower()
    processed_text = nltk.word_tokenize(processed_text)
    processed_text = [word for word in processed_text if word.isalnum()]
    processed_text = [word for word in processed_text if word not in nltk.corpus.stopwords.words('english')]
    return processed_text
```

```python
# apply the function
processed_docs = [' '.join(process_text(doc)) for doc in documents]
processed_docs
```

## TF-IDF

```python
# index the documents with Tf-Idf
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(processed_docs)
```

```python
# sample query
# query = "What is the tallest mountain in the world?"
# query = "What is located in the france"
query = "What is the tallest natural structure?"
```

```python
# try the sample query on our documents
query_vec = vectorizer.transform([query])
results = np.dot(tfidf_matrix, query_vec.T).toarray()
results
```

```python
# Build the function to search the document with Tf-Idf
def search(query, vectorizer, tfidf_matrix):
    query_vec = vectorizer.transform([query])
    results = np.dot(tfidf_matrix, query_vec.T).toarray()
    return results
```

```python
# Sort and display the function
results = search(query, vectorizer, tfidf_matrix)
results = sorted(enumerate(results), key = lambda x: x[1][0], reverse = True)
results
```

```python
print(f'Query: {query}\n')
for idx, score in results:
    print(f'Score: {score[0]:.2f} => Document: {documents[idx]}')
```
